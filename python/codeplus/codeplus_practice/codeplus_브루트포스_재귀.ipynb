## 부분수열의 합
https://www.acmicpc.net/problem/1182

```python
n, s = map(int, input().split())
arr = list(map(int, input().split()))
cnt = 0

for i in range(1, 1 << n):
    t = 0
    for j in range(n):
        if i & (1 << j):
            t += arr[j]
    if t == s:
        cnt += 1
print(cnt)
```

```python
n, s = map(int, input().split())
arr = list(map(int, input().split()))
cnt = 0
def permutation(idx, nums):
    global cnt
    if idx == n:
        if nums == s:
            cnt += 1
        return
    
    permutation(idx+1, nums+arr[idx])
    permutation(idx+1, nums)
    
permutation(0, 0)
if s == 0:
    cnt -= 1
print(cnt)
```

## 부분수열의 합

```python
n = int(input())
ans = [True] * 2000001
arr = list(map(int, input().split()))

for a in arr:
    ans[a] = False

def permutation(idx, nums):
    if idx == n:
        ans[nums] = False
        return
    
    permutation(idx+1, nums+arr[idx])
    permutation(idx+1, nums)

permutation(0, 0)

for i in range(1, 20000001):
    if ans[i]:
        print(i)
        break
```

## 연산자 끼워넣기
https://www.acmicpc.net/problem/14888

```python
n = int(input())
nums = list(map(int, input().split()))
arr = list(map(int, input().split()))
operate = []
for i, op in enumerate(arr):
    for _ in range(op):
        if i == 0:
            operate.append('+')
        elif i == 1:
            operate.append('-')
        elif i == 2:
            operate.append('*')
        else:
            operate.append('/')
check = [False] * len(operate)

def calc(idx, s, op):
    if op == '+':
        return s + nums[idx]
    elif op == '-':
        return s - nums[idx]
    elif op == '*':
        return s * nums[idx]
    else:
        if s >= 0:
            return s // nums[idx]
        else:
            return -(-s // nums[idx])

ans = []
def back_tracking(idx, s):
    global ans
    if idx == n:
        ans.append(s)
        return
    for i in range(len(operate)):
        if check[i]:
            continue
        check[i] = True
        back_tracking(idx+1, calc(idx, s, operate[i]))
        check[i] = False
        
back_tracking(1, nums[0])
ans.sort()
print(ans[-1])
print(ans[0])
```

```python
n = int(input())
nums = list(map(int, input().split()))
plus, minus, mul, div = map(int, input().split())
ans = []

def calc(idx, cur, plus, minus, mul, div):
    global ans
    if idx == n:
        ans.append(cur)
        return
    if plus > 0:
        calc(idx+1, cur+nums[idx], plus-1, minus, mul, div)
    if minus > 0:
        calc(idx+1, cur-nums[idx], plus, minus-1, mul, div)
    if mul > 0:
        calc(idx+1, cur*nums[idx], plus, minus, mul-1, div)
    if div > 0:
        if cur >= 0:
            calc(idx+1, cur//nums[idx], plus, minus, mul, div-1)
        else:
            calc(idx+1, -(-cur//nums[idx]), plus, minus, mul, div-1)

calc(1, nums[0], plus, minus, mul, div)

print(sorted(ans))
```

## 연산자 끼워넣기2
https://www.acmicpc.net/problem/15658

```python
import sys 

n = int(input())
num = list(map(int, input().split()))
op_cnt = list(map(int, input().split()))
max_result = -sys.maxsize
min_result = sys.maxsize


    
def operator(index: int, s: int) -> list:
    if index == n:
        global max_result, min_result
        max_result = max(max_result, s)
        min_result = min(min_result, s)
        return 
   
    for i in range(len(op_cnt)):
        if not op_cnt[i] == 0:
            if i == 0:
                result = s + num[index]
            elif i == 1:
                result = s - num[index]
            elif i == 2:
                result = s * num[index]
            else:
                if s < 0:
                    result = -s // num[index]
                    result = -result
                else:
                    result = s // num[index]

            op_cnt[i] -= 1
            operator(index + 1, result)    
            op_cnt[i] += 1


operator(1, num[0])


print(max_result)
print(min_result)
```

## 테트로미노
https://www.acmicpc.net/problem/14500

```python
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
c = [[False]*m for _ in range(n)]

def go(x, y, s, cnt):
    if cnt == 4:
        global ans
        if ans < s:
            ans = s
        return
    if x < 0 or x >= n or y < 0 or y >= m:
        return
    if c[x][y]:
        return
    c[x][y] = True
    for k in range(4):
        go(x+dx[k], y+dy[k], s+a[x][y], cnt+1)
    c[x][y] = False

ans = 0

for i in range(n):
    for j in range(m):
        go(i, j, 0, 0)
        if j+2 < m:
            temp = a[i][j] + a[i][j+1] + a[i][j+2]
            if i-1 >= 0:
                temp2 = temp + a[i-1][j+1]
                if ans < temp2:
                    ans = temp2
            if i+1 < n:
                temp2 = temp + a[i+1][j+1]
                if ans < temp2:
                    ans = temp2
        if i+2 < n:
            temp = a[i][j] + a[i+1][j] + a[i+2][j]
            if j+1 < m:
                temp2 = temp + a[i+1][j+1]
                if ans < temp2:
                    ans = temp2
            if j-1 >= 0:
                temp2 = temp + a[i+1][j-1]
                if ans < temp2:
                    ans = temp2
print(ans)
```

## 두 동전
https://www.acmicpc.net/problem/16197

- 총 4개의 방향을 10번까지 수행 가능
- 방법의 수: 4 ^ 10

```python
n, m = map(int, input().split())
graph = [list(input()) for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
x1 = y1 = x2 = y2 = -1
for i in range(n):
    for j in range(m):
        if graph[i][j] == 'o':
            if x1 == -1:
                x1, y1 = i, j
            else:
                x2, y2 = i, j

def go(step, x1, y1, x2, y2):
    if step == 11:
        return -1
    fall1 = fall2 = False
    if x1 < 0 or x1 >= n or y1 < 0 or y1 >= m:
        fall1 = True
    if x2 < 0 or x2 >= n or y2 < 0 or y2 >= m:
        fall2 = True
    if fall1 and fall2:
        return -1
    if fall1 or fall2:
        return step
    ans = -1
    for k in range(4):
        nx1, ny1 = x1+dx[k], y1+dy[k]
        nx2, ny2 = x2+dx[k], y2+dy[k]

        if 0 <= nx1 < n and 0 <= ny1 < m and graph[nx1][ny1] == '#':
            nx1, ny1 = x1, y1
        if 0 <= nx2 < n and 0 <= ny2 < m and graph[nx2][ny2] == '#':
            nx2, ny2 = x2, y2
        temp = go(step+1, nx1, ny1, nx2, ny2)
        if temp == -1:
            continue
        if ans == -1 or ans > temp:
            ans = temp
    return ans 

print(go(0, x1, y1, x2, y2))
```

## 에너지 모으기
https://www.acmicpc.net/problem/16198

```python
n = int(input())
arr = list(map(int, input().split()))

def go(arr):
    n = len(arr)
    if n == 2:
        return 0
    ans = 0
    for i in range(1, n-1):
        energy = arr[i-1]*arr[i+1]
        b = arr[:i] + arr[i+1:]
        energy += go(b)
        if ans < energy:
            ans = energy
    return ans
    
print(go(arr))
```

## N-Queen
https://www.acmicpc.net/problem/9663

```python
def check(row, col):
    if check_col[col]:
        return False
    if check_dig[row+col]:
        return False
    if check_dig2[row-col+n-1]:
        return False
    return True

def calc(row):
    if row == n:
        return 1
    ans = 0
    for col in range(n):
        if check(row, col):
            check_dig[row+col] = True
            check_dig2[row-col+n-1]=True
            check_col[col] = True
            a[row][col] = True
            
            ans += calc(row+1)
            
            check_dig[row+col] = False
            check_dig2[row-col+n-1] = False
            check_col[col] = False
            a[row][col] = False
    return ans

n = int(input())
a = [[False]*n for _ in range(n)]
check_col = [False] * n
check_dig = [False] * (2*n-1)
check_dig2 = [False] * (2*n-1)
print(calc(0))
```

```python
n = int(input())
row = [0] * n
ans = 0

def check(r):
    for i in range(r):
        if row[r] == row[i] or abs(row[r] - row[i]) == abs(r - i):
            return False
    return True

def chess(r):
    global ans
    if r == n:
        ans += 1
    else:
        for i in range(n):
            row[r] = i
            if check(r):
                chess(r+1)
chess(0)
print(ans)
```

## 스도쿠
https://www.acmicpc.net/problem/2580

```python
def square(x, y):
    return (x//3)*3 + (y//3)

def go(z):
    if z == 81:
        for row in a:
            print(*row)
        return True
    x = z//n
    y = z%n
    if a[x][y] != 0:
        return go(z+1)
    else:
        for i in range(1, 10):
            if c[x][i] == False and c2[y][i] == False and c3[square(x,y)][i] == False:
                c[x][i] = c2[y][i] = c3[square(x, y)][i] = True
                a[x][y] = i
                if go(z+1):
                    return True
                a[x][y] = 0
                c[x][i] = c2[y][i] = c3[square(x, y)][i] = False
    return False
n = 9
a = [list(map(int, input().split())) for _ in range(n)]
c = [[False]*10 for _ in range(n)]
c2 = [[False]*10 for _ in range(n)]
c3 = [[False]*10 for _ in range(n)]
for i in range(n):
    for j in range(n):
        if a[i][j] != 0:
            c[i][a[i][j]] = True
            c2[j][a[i][j]] = True
            c3[square(i, j)][a[i][j]] = True

go(0)
```

## 스도미노쿠
https://www.acmicpc.net/problem/4574

```python
n = 9
dx = [0, 1]
dy = [1, 0]
def convert(s):
    return (ord(s[0])-ord('A'), ord(s[1])-ord('1'))
def square(x, y):
    return (x//3)*3 + (y//3)
def can(x, y, num):
    return not c[x][num] and not c2[y][num] and not c3[square(x, y)][num]
def check(x, y, num, what):
    c[x][num] = what
    c2[y][num] = what
    c3[square(x, y)][num] = what
def check_range(x, y):
    return 0 <= x < n and 0 <= y < n
def go(z):
    if z == 81:
        for i in range(n):
            print(*a[i])
        return True
    x = z//n
    y = z%n
    if a[x][y] != 0:
        return go(z+1)
    else:
        for k in range(2):
            nx, ny = x+dx[k], y+dy[k]
            if not check_range(nx, ny):
                continue
            if a[nx][ny] != 0:
                continue
            for i in range(1, 10):
                for j in range(1, 10):
                    if i == j:
                        continue
                    if domino[i][j]:
                        continue
                    if can(x, y, i) and can(nx, ny, j):
                        check(x, y, i, True)
                        check(nx, ny, j, True)
                        domino[i][j] = domino[j][i] = True
                        a[x][y] = i
                        a[nx][ny] = j
                        if go(z+1):
                            return True
                        check(x, y, i, False)
                        check(nx, ny, j, False)
                        domino[i][j] = domino[j][i] = False
                        a[x][y] = 0
                        a[nx][ny] = 0
    return False
tc = 1
while True:
    c = [[False]*10 for _ in range(10)]
    c2 = [[False]*10 for _ in range(10)]    
    c3 = [[False]*10 for _ in range(10)]    
    domino = [[False]*10 for _ in range(10)]
    a = [[0]*9 for _ in range(9)]
    m = int(input())
    if m == 0:
        break
    for i in range(m):
        n1, s1, n2, s2 = input().split()
        n1 = int(n1)
        n2 = int(n2)
        x1, y1 = convert(s1)
        x2, y2 = convert(s2)
        a[x1][y1] = n1
        a[x2][y2] = n2
        domino[n1][n2] = domino[n2][n1] = True
        check(x1, y1, n1, True)
        check(x2, y2, n2, True)
    temp = input().split()
    for i in range(1, 10):
        s = temp[i-1]
        x, y = convert(s)
        a[x][y] = i
        check(x, y, i, True)
    print('Puzzle %d' %tc)
    go(0)
    tc += 1
```
