## 부등호
https://www.acmicpc.net/problem/2529

```python
import sys
n = int(input())
operate = input().split()
check = [False] * 10
min_num = str(sys.maxsize)
max_num = str(-sys.maxsize)

def calc(op, a, b):
    if op == '<':
        if a > b:
            return False
    elif op == '>':
        if a < b:
            return False
    return True

def back_tracking(idx: int, num: str) -> None:
    global min_num, max_num
    if idx == n+1:
        if min_num > num:
            min_num = num
        if max_num < num:
            max_num = num
        return
    
    for i in range(10):
        if check[i]:
            continue
        check[i] = True
        if idx == 0 or calc(operate[idx-1], num[idx-1], str(i)):
            back_tracking(idx+1, num+str(i))
        check[i] = False
        
back_tracking(0, '')
print(max_num)
print(min_num)
```

## 단어 수학
https://www.acmicpc.net/problem/1339

```python
def prev_permutation(a):
    i = len(a) - 1
    while i > 0 and a[i-1] <= a[i]:
        i -= 1
    if i <= 0:
        return False
    j = len(a) - 1
    while a[j] >= a[i-1]:
        j -= 1
    a[i-1], a[j] = a[j], a[i-1]
    
    j = len(a) - 1
    while i < j:
        a[i], a[j] = a[j], a[i]
        i += 1
        j -= 1
        
    return True

def calc(a, letters, d):
    m = len(letters)
    ans = 0
    alpha = dict()
    for i in range(m):
        alpha[letters[i]] = d[i]
    for s in a:
        now = 0
        for x in s:
            now = now*10 + alpha[x]
        ans += now
    return ans

n = int(input())
a = [''] * n
letters = set()
for i in range(n):
    a[i] = input()
    letters |= set(a[i])
letters = list(letters)
m = len(letters)
d = [i for i in range(9, 9-m, -1)]
ans = 0
while True:
    now = calc(a, letters, d)
    if ans < now:
        ans = now
    if not prev_permutation(d):
        break
print(ans)
```

## 연산자 끼워넣기
https://www.acmicpc.net/problem/14888

```python
def next_permutation(a):
    i = len(a) - 1
    while i > 0 and a[i-1] >= a[i]:
        i -= 1
    if i == 0:
        return False
    
    j = len(a) - 1
    while a[i-1] >= a[j]:
        j -= 1
    a[i-1], a[j] = a[j], a[i-1]
    
    j = len(a) - 1
    while i < j:
        a[i], a[j] = a[j], a[i]
        i += 1
        j -= 1
    return True

def div(a, b):
    if a >= 0:
        return a // b
    else:
        return -(-a // b)

def calc(a, b):
    n = len(a)
    ans = a[0]
    for i in range(1, n):
        if b[i-1] == 0:
            ans = ans + a[i]
        elif b[i-1] == 1:
            ans = ans - a[i]
        elif b[i-1] == 2:
            ans = ans * a[i]
        else:
            ans = div(ans, a[i])
    return ans

n = int(input())
a = list(map(int, input().split()))
cnts = list(map(int, input().split()))
b = []
for i, cnt in enumerate(cnts):
    for k in range(cnt):
        b.append(i)
b.sort()
ans = []
while True:
    temp = calc(a, b)
    ans.append(temp)
    if not next_permutation(b):
        break

print(max(ans))
print(min(ans))
```

## 스타트와 링크
