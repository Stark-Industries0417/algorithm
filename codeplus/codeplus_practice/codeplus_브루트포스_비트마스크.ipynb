## 부분수열의 합
https://www.acmicpc.net/problem/14225

```python
n = int(input())
arr = list(map(int, input().split()))
check = [False] * 2000001

for i in range(1, 1<<n):
    s = 0
    for k in range(n):
        if i & (1<<k):
            s += arr[k]
    check[s] = True
for i in range(1, 2000001):
    if not check[i]:
        print(i)
        break
```

## 가르침
https://www.acmicpc.net/problem/1062

```python
# 시간초과

n, k = map(int, input().split())
words = [input() for _ in range(n)]
check = [False] * 26
check[ord("a") - 97] = check[ord("n") - 97] = check[ord("t") - 97] = check[
    ord("i") - 97
] = check[ord("c") - 97] = True
ans = 0


def check_alpha():
    cnt = 0
    for w in words:
        ok = True
        for x in w:
            if not check[ord(x) - 97]:
                ok = False
                break
        if ok:
            cnt += 1
    return cnt


def select_alpha(idx, start):
    global ans
    if idx == k - 5:
        num = check_alpha()
        if ans < num:
            ans = num
        return
    for i in range(start, ord("z") - 97 + 1):
        if check[i]:
            continue
        check[i] = True
        select_alpha(idx + 1, i + 1)
        check[i] = False


select_alpha(0, 0)
print(ans)
```

```python
n, k = map(int, input().split())
words = [0] * n
for i in range(n):
    for x in input():
        words[i] |= 1 << (ord(x) - 97)

check = [False] * 26
check[ord("a") - 97] = check[ord("n") - 97] = check[ord("t") - 97] = check[
    ord("i") - 97
] = check[ord("c") - 97] = True
num = 0
num |= 1 << 0
num |= 1 << (ord("n") - 97)
num |= 1 << (ord("t") - 97)
num |= 1 << (ord("i") - 97)
num |= 1 << (ord("c") - 97)

ans = 0


def check_alpha(mask):
    cnt = 0
    for word in words:
        if (word & ((1 << 26) - 1 - mask)) == 0:
            cnt += 1
    return cnt


def select_alpha(idx, start, mask):
    global ans
    if idx == k - 5:
        num = check_alpha(mask)
        if ans < num:
            ans = num
        return
    for i in range(start, ord("z") - 97 + 1):
        if check[i]:
            continue
        check[i] = True
        select_alpha(idx + 1, i + 1, mask | (1 << i))
        check[i] = False


select_alpha(0, 0, num)
print(ans)
```

```python
def alpha_count(mask):
    cnt = 0
    for word in words:
        if (word & ((1 << 26) - 1 - mask)) == 0:
            cnt += 1
    return cnt


def go(idx, k, mask):
    if k < 0:
        return 0
    if idx == 26:
        return alpha_count(mask)
    ans = 0
    t1 = go(idx + 1, k - 1, mask | (1 << idx))
    if ans < t1:
        ans = t1
    if idx not in [
        ord("a") - ord("a"),
        ord("n") - ord("a"),
        ord("t") - ord("a"),
        ord("i") - ord("a"),
        ord("c") - ord("a"),
    ]:
        t2 = go(idx + 1, k, mask)
        if ans < t2:
            ans = t2
    return ans


n, k = map(int, input().split())
words = [0] * n
for i in range(n):
    for x in input():
        words[i] |= 1 << (ord(x) - 97)

print(go(0, k, 0))
```

## 구슬 탈출2
https://www.acmicpc.net/problem/13460

```python
1 & 3
```

```python
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
LIMIT = 10


class Result:
    def __init__(self, moved, hole, x, y):
        self.moved = moved
        self.hole = hole
        self.x = x
        self.y = y


def gen(k):
    a = [0] * LIMIT
    for i in range(LIMIT):
        a[i] = k & 3
        k >>= 2
    return a


def simulate(a, k, x, y):
    n = len(a)
    m = len(a[0])
    if a[x][y] == ".":
        return Result(False, False, x, y)
    moved = False
    while True:
        nx, ny = x + dx[k], y + dy[k]
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            return Result(moved, False, x, y)
        ch = a[nx][ny]
        if ch == "#":
            return Result(moved, False, x, y)
        elif ch in "RB":
            return Result(moved, False, x, y)
        elif ch == ".":
            a[x][y], a[nx][ny] = a[nx][ny], a[x][y]
            x, y = nx, ny
            moved = True
        elif ch == "O":
            a[x][y] = "."
            moved = True
            return Result(moved, True, x, y)


def check(a, dirs):
    n = len(a)
    m = len(a[0])
    hx, hy = 0, 0
    rx, ry = 0, 0
    bx, by = 0, 0

    for i in range(n):
        for j in range(m):
            if a[i][j] == "O":
                hx, hy = i, j
            elif a[i][j] == "R":
                rx, ry = i, j
            elif a[i][j] == "B":
                bx, by = i, j
    cnt = 0
    for k in dirs:
        cnt += 1
        hole1 = hole2 = False
        while True:
            p1 = simulate(a, k, rx, ry)
            rx, ry = p1.x, p1.y
            p2 = simulate(a, k, bx, by)
            bx, by = p2.x, p2.y

            if not p1.moved and not p2.moved:
                break
            if p1.hole:
                hole1 = True
            if p2.hole:
                hole2 = True
        if hole2:
            return -1
        if hole1:
            return cnt
    return -1


def valid(dirs):
    l = len(dirs)
    for i in range(l - 1):
        if dirs[i] == 0 and dirs[i + 1] == 1:
            return False
        if dirs[i] == 1 and dirs[i + 1] == 0:
            return False
        if dirs[i] == 2 and dirs[i + 1] == 3:
            return False
        if dirs[i] == 3 and dirs[i + 1] == 2:
            return False
        if dirs[i] == dirs[i + 1]:
            return False
    return True


n, m = map(int, input().split())
original = [input() for _ in range(n)]
ans = -1
for k in range(1 << (LIMIT * 2)):
    dirs = gen(k)
    if not valid(dirs):
        continue
    a = [list(row) for row in original]
    cur = check(a, dirs)
    if cur == -1:
        continue
    if ans == -1 or ans > cur:
        ans = cur

print(ans)
```
