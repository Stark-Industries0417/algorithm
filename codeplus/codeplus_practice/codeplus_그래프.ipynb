## Two Dots
https://www.acmicpc.net/problem/16929

```python
def cycle(color, x, y, cnt, start_x, start_y):
    global ans
    if ans:
        return

    for k in range(4):
        nx, ny = x + dx[k], y + dy[k]
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        if cnt >= 4 and nx == start_x and ny == start_y:
            ans = True
            return
        if graph[nx][ny] == color and not visited[nx][ny]:
            visited[nx][ny] = True
            cycle(color, nx, ny, cnt + 1, start_x, start_y)
            visited[nx][ny] = False


def game_start():
    for i in range(n):
        for j in range(m):
            visited[i][j] = True
            cycle(graph[i][j], i, j, 1, i, j)
            if ans:
                return "Yes"
    return "No"


n, m = map(int, input().split())
graph = [list(input()) for _ in range(n)]
visited = [[False] * m for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
ans = False

print(game_start())
```

## 서울 지하철 2호선
https://www.acmicpc.net/problem/16947

```python
import sys
from collections import deque

sys.setrecursionlimit(10 ** 6)

def dfs(x, cnt):
    if check[x]:
        if cnt - dist[x] >= 3:
            return x
        else:
            return -1
    check[x] = 1
    dist[x] = cnt
    for y in graph[x]:
        start = dfs(y, cnt + 1)
        if start != -1:
            check[x] = 2
            if x == start:
                return -1
            else:
                return start
    return -1


n = int(input())
check = [0] * (n + 1)
dist = [0] * (n + 1)
graph = [[] for _ in range(n + 1)]
for _ in range(n):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

dfs(1, 0)

q = deque()
for i in range(1, n + 1):
    if check[i] == 2:
        q.append(i)
        dist[i] = 0
    else:
        dist[i] = -1
while q:
    x = q.popleft()
    for y in graph[x]:
        if dist[y] == -1:
            q.append(y)
            dist[y] = dist[x] + 1

print(*dist[1:])
```

## 육각 보드
https://www.acmicpc.net/problem/12946

```python
import sys

sys.setrecursionlimit(10 ** 6)


def dfs(x, y, c):
    global ans
    color[x][y] = c
    ans = max(ans, 1)
    for k in range(6):
        nx, ny = x + dx[k], y + dy[k]
        if 0 <= nx < n and 0 <= ny < n:
            if a[nx][ny] == "X":
                if color[nx][ny] == -1:
                    dfs(nx, ny, 1 - c)
                ans = max(ans, 2)
                if color[nx][ny] == c:
                    ans = max(ans, 3)


n = int(input())
a = [input() for _ in range(n)]
color = [[-1] * n for _ in range(n)]
dx = [-1, -1, 0, 0, 1, 1]
dy = [0, 1, -1, 1, -1, 0]
ans = 0

for i in range(n):
    for j in range(n):
        if a[i][j] == "X" and color[i][j] == -1:
            dfs(i, j, 0)

print(ans)
```
