## 부등호
https://www.acmicpc.net/problem/2529

```python
n = int(input())
sign = input().split()
check = [False] * 10
ans = []

def right_sign(x, y, op):
    if op == '<':
        if x > y:
            return False
    else:
        if x < y:
            return False
    return True


def back_tracking(idx, s):
    if idx == n+1:
        ans.append(s)
        return
    
    for i in range(10):
        if check[i]:
            continue
        if idx == 0 or right_sign(s[idx-1], str(i), sign[idx-1]):
            check[i] = True
            back_tracking(idx+1, s + str(i))
            check[i] = False

back_tracking(0, '')
ans.sort()

print(ans[-1])
print(ans[0])
```

## nm과 k
https://www.acmicpc.net/problem/18290

```python
n, m, k = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
check = [[False] * m for _ in range(n)]
ans = -int(1e9)

dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]

def go(px, py, cnt, s):
    global ans
    if cnt == k:
        if ans < s:
            ans = s
        return
    
    for x in range(px, n):
        for y in range(py if x == px else 0, m):
            if check[x][y]:
                continue
            ok = True
            for i in range(4):
                nx, ny = x+dx[i], y+dy[i]
                if 0 <= nx < n and 0 <= ny < m:
                    if check[nx][ny]:
                        ok = False
            if ok:
                check[x][y] = True
                go(x, y, cnt+1, s+a[x][y])
                check[x][y] = False
go(0, 0, 0, 0)
print(ans)
```

## 단어 수학
https://www.acmicpc.net/problem/1339

```python
n = int(input())
words = [input() for _ in range(n)]
alpha = {}

for i in range(len(words)):
    for j in range(len(words[i])):
        if words[i][j] in alpha:
            alpha[words[i][j]] += 10 ** (len(words[i]) - j - 1)
        else:
            alpha[words[i][j]] = 10 ** (len(words[i]) - j - 1)
            
numsList = sorted(alpha.values(), reverse=True)

total, pow = 0, 9
for nums in numsList:
    total += pow * nums
    pow -= 1

print(total)
```

## 연산자 끼워넣기
https://www.acmicpc.net/problem/14888

```python
n = int(input())
nums = list(map(int, input().split()))
operators = list(map(int, input().split()))
operation = ''

for i in range(len(operators)):
    if i == 0:
        operation += '+' * operators[i]
    elif i == 1:
        operation += '-' * operators[i]
    elif i == 2:
        operation += '*' * operators[i]
    else:
        operation += '/' * operators[i]

check = [False] * len(operation)

def calc(a, o, b):
    if o == '+':
        return a + b
    elif o == '-':
        return a - b
    elif o == '*':
        return a * b
    else:
        if a < 0:
            return -(-a // b)
        return a // b

ans = []
def go(idx, s):
    global ans
    if idx == n:
        ans.append(s)
        return
    
    for i in range(len(operation)):
        if check[i]:
            continue
        check[i] = True
        now = calc(s, operation[i], nums[idx])
        go(idx+1, now)
        check[i] = False

go(1, nums[0])

print(max(ans))
print(min(ans))
```

## 스타트와 링크
https://www.acmicpc.net/problem/14889

```python
n = int(input())
ability = [list(map(int, input().split())) for _ in range(n)]
ans = 1e9

def go(idx, first, second):
    if idx == n:
        global ans

        if len(first) != n//2:
            return
        if len(second) != n//2:
            return
        t1 = t2 = 0
        for i in range(n//2):
            for j in range(n//2):
                if i == j:
                    continue
                t1 += ability[first[i]][first[j]]
                t2 += ability[second[i]][second[j]]
        diff = abs(t1 - t2)
        if ans > diff:
            ans = diff
        return
    
    if idx > n:
        return
    
    go(idx+1, first + [idx], second)
    go(idx+1, first, second + [idx])

go(0, [], [])

print(ans)
```

## 종이조각
https://www.acmicpc.net/problem/14391

```python
n, m = map(int, input().split())
paper = [list(map(int, input())) for _ in range(n)]
ans = -1e9
for p in range(1 << n*m):
    s = 0
    for i in range(n):
        cur = 0
        for j in range(m):
            k = i*m + j
            if p & (1 << k) == 0:
                cur = cur * 10 + paper[i][j]
            else:
                s += cur
                cur = 0
        s += cur
    
    for j in range(m):
        cur = 0
        for i in range(n):
            k = i*m + j
            if p & (1 << k):
                cur = cur * 10 + paper[i][j]
            else:
                s += cur
                cur = 0
        s += cur
    
    ans = max(ans, s)

print(ans)
```

## 테트로미노

```python
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]
check = [[False] * m for _ in range(n)]
ans = 0

def go(x, y, s, cnt):
    if cnt == 4:
        global ans
        if ans < s:
            ans = s
        return

    if x < 0 or y  < 0 or x >= n or y >= m:
        return
    
    if check[x][y]:
        return
    check[x][y] = True
    for k in range(4):
        go(x+dx[k], y+dy[k], s+a[x][y], cnt+1)
    check[x][y] = False

for i in range(n):
    for j in range(m):
        go(i, j, 0, 0)
        if j+2 < m:
            temp = a[i][j] + a[i][j+1] + a[i][j+2]
            if i-1 >= 0:
                temp2 = temp + a[i-1][j+1]
                ans = max(ans, temp2)
            if i+1 < n:
                temp2 = temp + a[i+1][j+1]
                ans = max(ans, temp2)
        if i+2 < n:
            temp = a[i][j] + a[i+1][j] + a[i+2][j]
            if j-1 >= 0:
                temp2 = temp + a[i+1][j-1]
                ans = max(ans, temp2)
            if j+1 < m:
                temp2 = temp + a[i+1][j+1]
                ans = max(ans, temp2)
print(ans)
```

## 두 동전
https://www.acmicpc.net/problem/16197

```python
n, m = map(int, input().split())
graph = [list(input()) for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
x1 = x2 = -1

for i in range(n):
    for j in range(m):
        if graph[i][j] == 'o':
            if x1 == -1:
                x1, y1 = i, j
            else:
                x2, y2 = i, j

def go(step, x1, y1, x2, y2):
    if step == 11:
        return -1
    fall1 = fall2 = False

    if x1 < 0 or x1 >= n or y1 < 0 or y1 >= m:
        fall1 = True
    if x2 < 0 or x2 >= n or y2 < 0 or y2 >= m:
        fall2 = True

    if fall1 and fall2:
        return -1
    if fall1 or fall2:
        return step

    ans = -1

    for k in range(4):
        nx1, ny1 = x1+dx[k], y1+dy[k]
        nx2, ny2 = x2+dx[k], y2+dy[k]

        if 0 <= nx1 < n and 0 <= ny1 < m and graph[nx1][ny1] == '#':
            nx1, ny1 = x1, y1
        if 0 <= nx2 < n and 0 <= ny2 < m and graph[nx2][ny2] == '#':
            nx2, ny2 = x2, y2
        
        temp = go(step+1, nx1, ny1, nx2, ny2)
        if temp == -1:
            continue

        if ans == -1 or ans > temp:
            ans = temp
    return ans

print(go(0, x1, y1, x2, y2))
```

## 에너지 모으기
https://www.acmicpc.net/problem/16198

```python
n = int(input())
e_li = list(map(int, input().split()))
ans = 0

def go(e_li, s):
    if len(e_li) == 2:
        global ans
        ans = max(ans, s)
        return

    for i in range(1, len(e_li)-1):
        energy = e_li[i-1] * e_li[i+1]
        b = e_li[:i] + e_li[i+1:]
        go(b, s+energy)

go(e_li, 0)

print(ans)
```

```python
n = int(input())
arr = list(map(int, input().split()))

def go(arr):
    n = len(arr)
    if n == 2:
        return 0
    ans = 0
    for i in range(1, n-1):
        energy = arr[i-1]*arr[i+1]
        b = arr[:i] + arr[i+1:]
        energy += go(b)
        if ans < energy:
            ans = energy
    return ans
    
print(go(arr))
```

## N-Queen
https://www.acmicpc.net/problem/9663

```python
n = int(input())
row = [0] * n
ans = 0

def check(r):
    for i in range(r):
        if row[r] == row[i] or abs(row[r] - row[i]) == abs(r - i):
            return False
    return True

def go(idx):
    if idx == n:
        global ans
        ans += 1
    else:
        for i in range(n):
            row[idx] = i
            if check(idx):
                go(idx+1)

go(0)
print(ans)
```

## 퇴사
https://www.acmicpc.net/problem/14501

```python
n = int(input())
day = [0] * (n + 1)
pay = [0] * (n + 1)
for i in range(1, n+1):
    a, b = list(map(int, input().split()))
    day[i] = a
    pay[i] = b

ans = 0

def go(idx, total):
    if idx == n+1:
        global ans
        ans = max(ans, total)
        return
    
    if idx > n+1:
        return
    
    go(idx+day[idx], total+pay[idx])
    go(idx+1, total)

go(1, 0)

print(ans)
```

## 구슬탈출2
https://www.acmicpc.net/problem/13460

```python
# 틀림

n, m = map(int, input().split())
board = [list(input()) for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]

for i in range(n):
    for j in range(m):
        if board[i][j] == 'R':
            rx, ry = i, j
        elif board[i][j] == 'B':
            bx, by = i, j

def move(rx, ry, bx, by, idx, cnt):
    fall1 = fall2 = False
    if idx == 10:
        return
    
    for k in range(4):
        nrx, nry = rx, ry
        while True:
            nrx += dx[k]
            nry += dy[k]
            if board[nrx][nry] == '#':
                nrx -= dx[k]
                nry -= dy[k]
                break
            if board[nrx][nry] == 'O':
                fall1 = True
                break
        nbx, nby = bx, by
        while True:
            nbx += dx[k]
            nby += dy[k]
            if board[nbx][nby] == '#':
                nbx -= dx[k]
                nby -= dy[k]
                break
            if board[nbx][nby] == 'O':
                fall2 = True
                break
        if nrx == nbx and nry == nby:
            if abs(nrx - rx) + abs(nry - ry) > abs(nbx - bx) + abs(nby - by):
                nrx -= dx[k]
                nry -= dy[k]
            else:
                nbx -= dx[k]
                nby -= dy[k]
        if fall1 and not fall2:
            return cnt
        if fall1 and fall2:
            return
        if not fall1 and fall2:
            return
        move(nrx, nry, nbx, nby, idx+1, cnt+1)
    return -1

print(move(rx, ry, bx, by, 0, 1))
```

```python
import sys
from copy import deepcopy
input = sys.stdin.readline

n = int(input())
board = [list(map(int, input().split())) for _ in range(n)]

def up(board):
    for j in range(n):
        pointer = 0
        for i in range(1, n):
            if board[i][j]:
                tmp = board[i][j]
                board[i][j] = 0
                if board[pointer][j] == 0:
                    board[pointer][j] = tmp
                elif board[pointer][j] == tmp:
                    board[pointer][j] *= 2
                    pointer += 1
                else:
                    pointer += 1
                    board[pointer][j] = tmp
    return board

def down(board):
    for j in range(n):
        pointer = n-1
        for i in range(n-2, -1, -1):
            if board[i][j]:
                tmp = board[i][j]
                board[i][j] = 0
                if board[pointer][j] == 0:
                    board[pointer][j] = tmp
                elif board[pointer][j] == tmp:
                    board[pointer][j] *= 2
                    pointer -= 1
                else:
                    pointer -= 1
                    board[pointer][j] = tmp
    return board

def left(board):
    for i in range(n):
        pointer = 0
        for j in range(1, n):
            if board[i][j]:
                tmp = board[i][j]
                board[i][j] = 0
                if board[i][pointer] == 0:
                    board[i][pointer] = tmp
                elif board[i][pointer] == tmp:
                    board[i][pointer] *= 2
                    pointer += 1
                else:
                    pointer += 1
                    board[i][pointer] = tmp
    return board

def right(board):
    for i in range(n):
        pointer = n-1
        for j in range(n-2, -1, -1):
            if board[i][j]:
                tmp = board[i][j]
                board[i][j] = 0
                if board[i][pointer] == 0:
                    board[i][pointer] = tmp
                elif board[i][pointer] == tmp:
                    board[i][pointer] *= 2
                    pointer -= 1
                else:
                    pointer -= 1
                    board[i][pointer] = tmp
    return board

def dfs(board, cnt):
    if cnt == 5:
        return max(map(max, board))
    
    return max(dfs(up(deepcopy(board)), cnt+1), dfs(down(deepcopy(board)), cnt+1), dfs(left(deepcopy(board)), cnt+1), dfs(right(deepcopy(board)), cnt+1))

print(dfs(board, 0))
```

## DSLR

```python
# 메모리 초과

from collections import deque

n = int(input())

def D(a):
    a = int(a) * 2
    if a > 9999:
        a %= 10000
    return str(a)

def S(a):
    a = int(a)
    if a == 0:
        a = 9999
    return str(a-1)

def L(a):
    return str(int(a[1:] + a[0]))

def R(a):
    return str(int(a[-1] + a[:-1]))

func = [D, S, L, R]

def bfs(s, e):
    q = deque([(s, '')])

    while q:
        x, move = q.popleft()
        if x == e:
            print(move)
            return
        for f in func:
            q.append((f(x), move + f.__name__))

for _ in range(n):
    start, end = input().split()
    bfs(start, end)
```

## 연구소
https://www.acmicpc.net/problem/14502

```python
from copy import deepcopy
from collections import deque

n, m = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]
ans = 0

vir = deque()
for i in range(n):
    for j in range(m):
        if board[i][j] == 2:
            vir.append((i, j))

def bfs(b):
    q = deepcopy(vir)
    while q:
        x, y = q.popleft()

        for k in range(4):
            nx, ny = x+dx[k], y+dy[k]
            if 0 <= nx < n and 0 <= ny < m and b[nx][ny] == 0:
                b[nx][ny] = 2
                q.append((nx, ny))
    zero = 0
    for i in range(n):
        zero += b[i].count(0)

    return zero
    
def go(idx, cnt):
    if cnt == 3:
        global ans
        safe = bfs(deepcopy(board))
        if ans < safe:
            ans = safe
        return

    if idx >= n*m:
        return

    x = idx // m
    y = idx % m

    if board[x][y] == 0:
        board[x][y] = 1
        go(idx+1, cnt+1)
        board[x][y] = 0

    go(idx+1, cnt)

go(0, 0)

print(ans)
```

```python
import sys
sys.setrecursionlimit(10 ** 6)

x, y, z = map(int, input().split())
check = [[False] * 1501 for _ in range(1501)]
s = x + y + z

def go(x, y):
    if check[x][y]:
        return
    check[x][y] = True
    a = [x, y, s - x -y]
    for i in range(3):
        for j in range(3):
            if a[i] < a[j]:
                b = [x, y, s - x - y]
                b[i] += a[i]
                b[j] -= a[j]
                go(b[0], b[1])

if s % 3 != 0:
    print(0)
else:
    go(x, y)
    if check[s // 3][s //3]:
        print(1)
    else:
        print(0)
```

## 아기 상어

```python
from collections import deque

n = int(input())
graph = [list(map(int, input().split())) for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]

def bfs(i, j, size):
    q = deque([(i, j)])
    d = [[-1] * n for _ in range(n)]
    d[i][j] = 0
    ans = []

    while q:
        x, y = q.popleft()
        for k in range(4):
            nx, ny = x+dx[k], y+dy[k]
            ok = eat = False
            if 0 <= nx < n and 0 <= ny < n and d[nx][ny] == -1:
                if graph[nx][ny] == 0:
                    ok = True
                elif graph[nx][ny] < size:
                    ok = eat = True
                elif graph[nx][ny] == size:
                    ok = True

            if ok:
                q.append((nx, ny))
                d[nx][ny] = d[x][y] + 1
                if eat:
                    ans.append((d[nx][ny], nx, ny))
    if not ans:
        return None
    ans.sort()
    return ans[0]

for i in range(n):
    for j in range(n):
        if graph[i][j] == 9:
            x, y = i, j
            graph[i][j] = 0

size = 2
exp = 0
answer = 0
while True:
    s = bfs(x, y, size)
    if s is None:
        break
    dist, x, y = s
    graph[x][y] = 0
    answer += dist
    exp += 1

    if exp == size:
        exp = 0
        size += 1

print(answer)
```

## 소수 경로

```python
from collections import deque

n = 10000
prime = [False, False] + [True] * 9998

for i in range(2, int(n ** 0.5) + 1):
    if prime[i]:
        for j in range(i**2, n, i):
            prime[j] = False

def change(num, idx, digit) -> int:
    if idx == 0 and digit == 0:
        return -1
    num = list(str(num))
    num[idx] = str(digit)
    return int(''.join(num))

for _ in range(int(input())):
    a, b = map(int, input().split())
    c = [False] * n
    dist = [0] * n
    c[a] = True
    dist[a] = 0

    q = deque([a])
    while q:
        now = q.popleft()
        for i in range(4):
            for j in range(10):
                nxt = change(now, i, j)
                if nxt != -1:
                    if prime[nxt] and not c[nxt]:
                        q.append(nxt)
                        dist[nxt] = dist[now] + 1
                        c[nxt] = True
    print(dist[b])
```

```python
from collections import deque

n = int(input())
graph = [list(input()) for _ in range(n)]
dx, dy = [0, 0, 1, -1], [1, -1, 0, 0]

def can(blind, c1, c2):
    if c1 == c2:
        return True
    if blind:
        if c1 == 'R' and c2 == 'G':
            return True
        if c1 == 'G' and c2 == 'R':
            return True
    return False


def bfs(blind):
    visited = [[False] * n for _ in range(n)]
    ans = 0
    for i in range(n):
        for j in range(n):
            if visited[i][j]:
                continue
            ans += 1
            q = deque()
            q.append((i, j))
            while q:
                x, y = q.popleft()
                for k in range(4):
                    nx, ny = x+dx[k], y+dy[k]
                    if 0 <= nx < n and 0 <= ny < n and not visited[nx][ny]:
                        if can(blind, graph[x][y], graph[nx][ny]):
                            visited[nx][ny] = True
                            q.append((nx, ny))
    return ans


print(bfs(False), bfs(True))
```

## 4연산
https://www.acmicpc.net/problem/14395

```python
from collections import deque

s, t = map(int, input().split())
limit = 10 ** 9

def bfs(s, t):
    if s == t:
        return 0
    q = deque([(s, '')])
    oper = ['*', '+', '-', '/']
    check = set()
    check.add(s)

    while q:
        num, op = q.popleft()
        if num == t:
            return op
        
        for k in oper:
            if k == '*':
                temp = num * num
            elif k == '+':
                temp = num + num
            elif k == '-':
                temp = num - num
            else:
                if num != 0:
                    temp = num / num
            if 0 <= temp <= limit and temp not in check:
                check.add(temp)
                q.append((temp, op+k))
    return -1

print(bfs(s, t))
```

## 행렬
https://www.acmicpc.net/problem/1080

```python
n, m = map(int, input().split())
A = [list(map(int, input())) for _ in range(n)]
B = [list(map(int, input())) for _ in range(n)]

def reverse(i, j):
    for x in range(i, i+3):
        for y in range(j, j+3):
            A[x][y] = 1 - A[x][y]

def check():
    for i in range(n):
        for j in range(m):
            if A[i][j] != B[i][j]:
                return False
    return True

ans = 0
for i in range(n-2):
    for j in range(m-2):
        if A[i][j] != B[i][j]:
            ans += 1
            reverse(i, j)

if check():
    print(ans)
else:
    print(-1)
```

## 전구와 스위치
https://www.acmicpc.net/problem/2138

```python
import sys
input = sys.stdin.readline

n = int(input().rstrip())
a = list(map(int, input().rstrip()))
B = list(map(int, input().rstrip()))

def change(idx):
    if idx == 0:
        for i in range(idx+2):
            A[i] = 1 - A[i]
    elif idx == n-1:
        for i in range(n-2, n):
            A[i] = 1 - A[i]
    else:
        for i in range(idx-1, idx+2):
            A[i] = 1 - A[i]

for i in range(2):
    A = a[:]
    cnt = 0
    for j in range(n):
        if j == 0:
            if i == 0 and A != B:
                cnt += 1
                change(j)

        elif 1 <= j <= n-2:
            if A[j-1] != B[j-1]:
                cnt += 1
                change(j)

        elif j == n-1:
            if A[j-1] != B[j-1]:
                cnt += 1
                change(j)

    if A == B:
        print(cnt)
        break
if A != B:
    print(-1)
```

## 동전 뒤집기
https://www.acmicpc.net/problem/1285

```python
n = int(input())
coin = [list(input()) for _ in range(n)]
res = n ** 2 + 1

for bit in range(1<<n):    
    tmp = [coin[i][:] for i in range(n)]

    for i in range(n):
        if bit & (1 << i):
            for j in range(n):
                if tmp[i][j] == 'H':
                    tmp[i][j] = 'T'
                else:
                    tmp[i][j] = 'H'
    
    tsum = 0
    for j in range(n):
        cnt = 0
        for i in range(n):
            if tmp[i][j] == 'T':
                cnt += 1
        tsum += min(cnt, n-cnt)
    res = min(res, tsum)

print(res)
```

## 보석도둑
https://www.acmicpc.net/problem/1202

```python
import heapq, sys
input = sys.stdin.readline

n, k = map(int, input().split())
jewels = []
for _ in range(n):
    m, v = list(map(int, input().split()))
    heapq.heappush(jewels, (m, v))

bags = [int(input()) for _ in range(k)]
bags.sort()

ans = 0
tmp = []
for bag in bags:
    while jewels and bag >= jewels[0][0]:
        m, v = heapq.heappop(jewels)
        heapq.heappush(tmp, -v)
    if tmp:
        ans += heapq.heappop(tmp)

print(-ans)
```

## 순회강연
https://www.acmicpc.net/problem/2109

```python
import heapq

n=int(input())
lst=[]
for i in range(n):
    lst.append(list(map(int, input().split())))

lst.sort(key=lambda x: (x[1]))
p_list=[]
for i in lst:
    heapq.heappush(p_list, i[0])
    if (len(p_list)>i[1]):
        heapq.heappop(p_list)

print(sum(p_list))
```

## 잃어버린 괄호
https://www.acmicpc.net/problem/1541

```python
import re

exp = input()
exp = re.split('([^0-9])', exp)
ops = [('+', '-') , ('-', '+')]

def calc(x, op, y):
    if op == '+':
        return int(x) + int(y)
    else:
        return int(x) - int(y)

ans = 1e9
for op in ops:
    stack = []
    cp = exp[:]
    for o in op:
        while cp:
            tmp = cp.pop(0)
            if tmp == o:
                stack.append(str(calc(stack.pop(), tmp, cp.pop(0))))
            else:
                stack.append(tmp)
        cp = stack
        stack = []
    if ans > int(cp[0]):
        ans = int(cp[0])
print(ans)
```

## 수 묶기
https://www.acmicpc.net/problem/1744

```python
n = int(input())
pos, neg, one = [], [], []
result = 0

for _ in range(n):
    num = int(input())
    if num > 1:
        pos.append(num)
    elif num <= 0:
        neg.append(num)
    else:
        one.append(num)

pos.sort(reverse=True)
neg.sort()

result += sum(pos[i] * pos[i+1] for i in range(0, len(pos), 2) if i+1 < len(pos))
if len(pos) % 2 != 0:
    result += pos[-1]

result += sum(neg[i] * neg[i+1] for i in range(0, len(neg), 2) if i+1 < len(neg))
if len(neg) % 2 != 0:
    result += neg[-1]

result += sum(one)

print(result)
```

## AB
https://www.acmicpc.net/problem/12970

```python
n, k = map(int, input().split())

s = list('B' * n)

def check():
    cnt = 0
    for i in range(n-1):
        if s[i] == 'A':
            for j in range(i+1, n):
                if s[j] == 'B':
                    cnt += 1
    return cnt

for i in range(n):
    s[i] = 'A'
    cnt = check()
    if cnt == k:
        break
    elif cnt > k:
        s[i] = 'B'

tmp = ''.join(s)
if tmp == 'A'*n or tmp == 'B'*n:
    if k == 0:
        print(tmp)
    else:
        print(-1)
else:
    print(tmp)
```

## A와 B
https://www.acmicpc.net/problem/12904

```python
s = list(input())
t = list(input())

ans = 0
while t:
    if t[-1] == 'A':
        t.pop()
    
    elif t[-1] == 'B':
        t.pop()
        t.reverse()
    
    if t == s:
        ans = 1
        break

print(ans)
```

## 숫자카드
https://www.acmicpc.net/problem/10815

```python
n = int(input())
card = list(map(int, input().split()))
m = int(input())
has = list(map(int, input().split()))

card.sort()

def binary_search(num):
    start, end = 0, len(card)-1

    while start <= end:
        mid = (start + end) // 2
        if card[mid] == num:
            return 1
        elif num < card[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return 0

for h in has:
    print(binary_search(h), end=' ')
```

## 종이의 개수
https://www.acmicpc.net/problem/1780

```python
n = int(input())
paper = list(list(map(int, input().split())) for _ in range(n))
minus, zero, one = 0, 0, 0

def search(x, y, n):
    global minus, zero, one
    check = paper[x][y]

    for i in range(x, x + n):
        for j in range(y, y + n):
            if paper[i][j] != check:
                for k in range(3):
                    for l in range(3):
                        search(x + k * n//3, y + l * n//3, n//3)
                return
    
    if check == -1:
        minus += 1
    elif check == 0:
        zero += 1
    else:
        one += 1

search(0, 0, n)
print(f'{minus}\n{zero}\n{one}')
```
